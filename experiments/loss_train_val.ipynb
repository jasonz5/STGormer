#### 绘制训练过程的train/val loss变化图

```python
import pickle
import matplotlib.pyplot as plt

file_path = 'NYCBike1/MoESoft-layers/STSTST/stats.pkl'
with open(file_path, 'rb') as f:
    data = pickle.load(f)
# for key in data.keys():
#     print(key, data[key].shape)

epochs = data['epoch']
train_loss = data['train_loss']
val_loss = data['val_loss']
learning_rates = data['cur_lr']
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plotting train_loss
axs[0].plot(epochs, train_loss, 'tab:blue')
axs[0].set_title('Train Loss')
axs[0].set_xlabel('Epochs')
axs[0].set_ylabel('Train Loss')
axs[0].grid(True)
    
# Plotting val_loss
axs[1].plot(epochs, val_loss, 'tab:orange')
axs[1].set_title('Validation Loss')
axs[1].set_xlabel('Epochs')
axs[1].set_ylabel('Validation Loss')
axs[1].grid(True)

# Plotting learning_rates
axs[2].plot(epochs, learning_rates, 'tab:green')
axs[2].set_title('Learning Rate')
axs[2].set_xlabel('Epochs')
axs[2].set_ylabel('Learning Rate')
axs[2].grid(True)

plt.tight_layout()
# plt.savefig('loss.png')
plt.show()
```
