```python
import numpy as np
file_paths = {
    'degree': 'degree.npy',
    's_in': 's_in.npy',
    'sa_bias': 'sa_bias.npy'
}
data = {}
for key, path in file_paths.items():
    try:
        data[key] = np.load(path)
        print(f"Shape of {key}: {data[key].shape}")
    except FileNotFoundError:
        print(f"Error: '{path}' not found.")
    except Exception as e:
        print(f"An error occurred while loading '{path}': {str(e)}")
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Heatmap for s_in
s_in_matrix = data['degree'].reshape(16, 8)
cmap_in = plt.get_cmap('RdBu')
norm_in = plt.Normalize(vmin=np.min(s_in_matrix), vmax=np.max(s_in_matrix))
im0 = axs[0].imshow(s_in_matrix, cmap=cmap_in, norm=norm_in)
axs[0].set_xticks(np.arange(-.5, 8, 1))
axs[0].set_yticks(np.arange(-.5, 16, 1))
axs[0].set_xticklabels([])
axs[0].set_yticklabels([])
axs[0].grid(which='major', color='black', linestyle='-', linewidth=1)
axs[0].tick_params(length=0)
cb0 = fig.colorbar(im0, ax=axs[0], orientation='vertical', shrink=0.6)

# Heatmap for sa_bias
node_id = 68
sa_bias_matrix = data['sa_bias'][node_id, :].reshape(16, 8)
norm_bias = plt.Normalize(vmin=np.min(sa_bias_matrix), vmax=np.max(sa_bias_matrix))
im1 = axs[1].imshow(sa_bias_matrix, cmap=cmap_in, norm=norm_bias)
axs[1].set_xticks(np.arange(-.5, 8, 1))
axs[1].set_yticks(np.arange(-.5, 16, 1))
axs[1].set_xticklabels([])
axs[1].set_yticklabels([])
axs[1].grid(which='major', color='black', linestyle='-', linewidth=1)
axs[1].tick_params(length=0)
cb1 = fig.colorbar(im1, ax=axs[1], orientation='vertical', shrink=0.6)
# Highlight the specific node with a red circle
highlight_node_row = node_id // 8
highlight_node_col = node_id % 8
axs[1].add_patch(plt.Circle((highlight_node_col, highlight_node_row), 0.65, color='red', fill=False, linewidth=2.5, zorder=3))

plt.tight_layout()
plt.show()
```
